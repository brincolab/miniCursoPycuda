##### Obtenemos una ecuacion de la forma radial en terminos adimensionales $\xi = r/r_0$ y $r_0 = \frac{\hbar}{2mU_0\rho}$ $$\\$$ $$ \frac{1}{\xi} \frac{d}{d\xi}\left(\xi \frac{d f(\xi)}{d\xi}\right) -\frac{f}{\xi^2} + f -f ^3 = 0$$

Esta ecuacion resulta ser no-lineal y no integrable con herramientas usuales usuales de computo. En realidad la manera que encontre es usando en el sentido mas lemental una serie de frobenius.

El problema lo planteo de la siguinete manera

Prime iteracion

```python
orden=55 #55
f0 = 0.0
f1 = 1.0 # Ojo, no he justificado por que debe ser esta la pendiente, un metodo de disparo seria interesante
f2 = 0.0
f3 = -0.5*f1
f4  = 2*f1
dx = 0.1
vortex = taylorDiff([f0,f1,f2,f3,f4],orden) # Polinomio solucion en la vecindad
```

Relacion de recurrencia. Libre de divergencias ... : o

```python
def gpeRRel(f , k): # k > 5
    return -1./(k-1) * ( f.jet[k-2] - (f**3).jet[k-2] + f.jet[k-1]*(k-1)*k )
```

```python
# Ahora si primera iteracion ... 
# OJO Truncamo en el orden 10 el calculo, ya que al elevar al cubo debemos tener suficiente espacio para el orden del polinomio
for k in range(5,7): 
    vortex.jet[k] = gpeRRel(vortex,k)
```

```python
def integStep(f,x,dx=0.01,ordenT=30,ordenTr=10):
    f0 = f(x)
    f1 = (f.deriv())(x) # Ojo, no he justificado por que debe ser esta la pendiente, un metodo de disparo seria interesante
    f2 = 0.0
    f3 = -0.5*f1
    f4  = 2*f1
    f = taylorDiff([f0,f1,f2,f3,f4],ordenT)
    for k in range(5,ordenTr): # OJO Truncamo en el orden 10 el calculo
        f.jet[k] = gpeRRel(f,k)
    #f
    f.draw(npoint=10,xdomain=[x-2*dx,x+2*dx])
    return f
```

```python
v = integStep(vortex,0.05,ordenTr=7)
integStep(v,0.1,ordenTr=7)
```

Iterando el metodo
---------
Hasta aqui conocemos la solucion en una vecindad de $x_0 = 0$, pero lo que queremos es conocerla en un dominio aun mayor. Lo que haremos simplemente es usar esta solucion y evaluarla en un punto cercano. Partiendo de este nuevo punto $x_1$ hacer de nuevo la recurrencia para conocer el polinomio solucion. Iterando el metodo podemos conocer la solucion.

```python
xDom = linspace(0.001, 0.1, 5000)
xDom
```

```python
vtxSolution = [vortex(0)]
for i  in xDom:
    f0 = vortex(i)
    f1 = (vortex.deriv())(i)
    f2 = 0.0
    f3 = -0.5*f1
    f4  = 2*f1
    vortex = taylorDiff([f0,f1,f2,f3,f4],orden)
    for k in range(5,7):
        vortex.jet[k] = gpeRRel(vortex,k)
    vtxSolution.append(vortex(i))
```
